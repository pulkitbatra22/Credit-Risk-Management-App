```python
import pandas as pd
from scipy.stats import chi2_contingency

# Assuming df is your DataFrame and 'dependent_var' is your dependent variable
df = cc_data_full_data

dependent_var = 'Is high risk'
alpha = 0.05

# Get list of categorical columns
categorical_columns = df.select_dtypes(include=['object', 'category']).columns.tolist()

# Remove the dependent variable from the list
categorical_columns.remove(dependent_var)

# Initialize a list to store the results
results = []

# Perform Chi-squared test for each categorical variable
for col in categorical_columns:
    contingency_table = pd.crosstab(df[dependent_var], df[col])
    chi2, p, dof, expected = chi2_contingency(contingency_table)
    result = {
        'Variable': col,
        'Min': df[col].min(),
        'Max': df[col].max(),
        'Count': df[col].count(),
        'Chi2': chi2,
        'p-value': p,
        'Reject Null Hypothesis': p < alpha
    }
    results.append(result)

# Convert results to DataFrame for better readability
results_df = pd.DataFrame(results)

# Print the results
print(results_df)
```

```python
# Univariate MAD outlier analysis 
# performing on high risk =0, as high risk =1 has too less observations already
```

```python
import pandas as pd
import numpy as np

df = cc_train_copy

# Specify numerical columns
numerical_cols_to_analyze = ['Income', 'Age', 'Family member count']

# Function to calculate MAD-based outliers
def is_outlier(s, threshold=3.5):
    median = np.median(s)
    diff = np.abs(s - median)
    mad = np.median(diff)
    if mad == 0:
        return np.zeros(s.shape, dtype=bool)  # Return all False if MAD is 0
    modified_z_score = 0.6745 * diff / mad
    return modified_z_score > threshold

# Split the DataFrame
high_risk_df = df[df['Is high risk'] == 1]
non_high_risk_df = df[df['Is high risk'] != 1]

# Identify outliers in the high risk subset
outliers_index = set()
for col in numerical_cols_to_analyze:
    outliers = is_outlier(non_high_risk_df[col])
    outliers_index.update(non_high_risk_df[outliers].index)

# Remove outliers from the high risk subset
non_high_risk_cleaned = non_high_risk_df.drop(index=outliers_index)

# Combine the cleaned high risk subset with the non-high risk subset
df_final = pd.concat([non_high_risk_cleaned, high_risk_df])

# Maintain the original order by sorting based on the index
df_final = df_final.sort_index()
```
